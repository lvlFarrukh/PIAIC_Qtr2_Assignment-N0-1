```python
import os
import numpy as np
import matplotlib.pylab as plt

img_name = []
        
def plti(im, h=8, **kwargs):
    """
    Helper function to plot an image.
    """
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none", **kwargs)
    plt.axis('on')
    #input('press <ENTER> to continue')
%matplotlib inline


root_dir = 'F:.\pics'
for dirName, subdirList, fileList in os.walk(root_dir):
    ar = np.zeros([len(fileList),200,200,3], dtype='int16')
    for c,fname in enumerate(fileList):
        im = plt.imread(f'{dirName}\{fname}')
        im = im[:200,:200,:3]
        ar[c] = im
        
plti(ar[5])
```
